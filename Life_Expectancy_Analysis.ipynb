```python
seed = np.arange(0,9)
for s in seed:
    np.random.seed(s)
    sample_1 = df[df["Life expectancy "].isin([70, 80])].sample(10).reset_index().drop(columns=["index"])
    mean.append(np.mean(sample_1["Life expectancy "]))
    std.append(np.std(sample_1["Life expectancy "]))

# Create a DataFrame to store the calculated values
samples = pd.DataFrame(columns=["mean", "std"], data=list(zip(mean, std)))
samples
```
