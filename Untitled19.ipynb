```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df[" income"]= lb.fit_transform(df[" income"])
```

```python
df
```

```python
categorical_data = df.columns[df.dtypes=="O"]
print(*categorical_data, sep=" | ")
```

```python
new_df = pd.get_dummies(df, columns=categorical_data)
new_df.head(10)
```

```python
x=new_df
y=new_df[" income"]
```

```python
x.shape
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
Gnb=GaussianNB()
Gnb.fit(x_train, y_train)
```

```python
predict1 = Gnb.predict(x_test)
print(predict1)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(predict1, y_test)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predict1)
cm
```

```python
sns.heatmap(cm, annot=True, cmap="Blues")
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
predict2=bnb.predict(x_test)
print(predict2)
```

```python
accuracy2 = accuracy_score(predict2, y_test)
accuracy2
```

```python
from sklearn.metrics import confusion_matrix
cm2 = confusion_matrix(predict2, y_test)
cm2
```

```python
sns.heatmap(cm2, annot=True, cmap="Blues")
```

```python
plt.barh(["Bernoulli", "Gaussion"], accuracy, accuracy2)
```
